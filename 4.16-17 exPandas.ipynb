#### 데이터프레임을 합치는데 사용하는 2개 메소드가 있는데,
#### 1. concat() : row 나 column 기준으로 단순하게 이어붙이기 (기준은 index 값)
#### 2. merge() :  특정 고유한 키 (unique id) 값을 기준으로 이어붙인다. - df, df2 붙일 때 어느정도 겹치는 기준이 있어야 함 예를 들어 이름. 

#### 붙이는 기준이 다른것을 기준으로 이어 붙이고 싶다 = merge()
#### 일반 인덱스 값 기준으로 붙이고 싶다 = concat()

#### df, df2 를 '이름' 기준으로 두 데이터프레임을 합치기

#### pd.merge(left, right, on = '기준 column', how = 'left')

```python
# left, right 방식 
df_right = df2.drop([1,3,5,7,9], axis=0)
```

```python
df_right # 1,3,5,7,9 날라감
```

```python
df_right = df_right.reset_index(drop=True)
```

```python
df_right
```

```python
# 이제 행들이 날라갔기 떄문에 비빌 구석은? merge 가 답임.

# 두 개의 데이터프레임을 합치는데, 합치는 기준 항목(열)은 
# '이름' 항목, 왼쪽 프레임의 '이름' 항목을 기준으로 순서로 정한다.

pd.merge(df, df_right, on = '이름', how = 'left') 
# how =  기준을 어디쪽으로 순서를 정렬할래? 두개가 항목은 동일한데 오른쪽? 왼쪽? 어디 기준으로 갈까요? 우리는 왼쪽으로 해보자.
# 예를 들어 슈가 가 왼쪽에선 14번인데 여기선 9번에 있으니까 어떤걸 기준으로 할래? 하는것임. 
```

```python
# 왼쪽은 15개, 오른쪽은 10개니까 nan 은 어차피 없으니까 빼버리고 걍 오른쪽 갯수에 맞춰버림. 
pd.merge(df, df_right, on = '이름', how = 'right') 
# 만약 13개 10개 이렇게 있는데 각각 없는 항목이 없는게 달라버리면 어쩔? inner!! 
```

```python
# inner, outer 방식
# inner : 두 개의 데이터프레임에 모두 키 값이 존재하는 경우만 합치기. 한두개 항목이라도 빠져있는 것은 합치지 말고!
# outer: 하나의 데이터프레임에 키값이 존재하기만 하면 모두 합치기
```

```python
df_left = df.drop([0,4], axis=0)
```

```python
df_left
```

```python
# inner : 두 개의 데이터프레임에 모두 키 값이 존재하는 경우만 합치기. (and 개념)
pd.merge(df_left, df_right, on = '이름', how = 'inner')

# 왼쪽 15-2 = 13개, 오른쪽 5개 지웠으니께 총 8개만 나옴. 
```
