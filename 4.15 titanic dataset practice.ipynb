### 수치형 결측치 데이터에 대한 처리

수치형 데이터의 결측치는 대체로 평균값을 구한뒤, 평균값으로 채워준다. 

```python
# 0으로 채웠을 때
# describe() 통계치 데이터를 보여준다
train['Age'].fillna(0).describe() # na: null 인 부분(비어있는)을 0으로 채운다. 
```

```python
# 평균 값으로 채웠을 때
meanAge = train['Age'].mean()
print(meanAge)

train['Age'].fillna(meanAge).describe()
```

```python
# 평균 값으로 채웠을 때
train['Age'].fillna(train['Age'].mean()).describe() # 하나의 열에 있는 데이터만 처리
# 1. fillna 안의 것을 돌리고 2. mean 값이 나오면 fillna 가 돌아가고 (train에 대해서) 3. describe 한다 
```

### scikit-learn 에서 imputer 라는 method 를 활용하면 
### 2개 이상의 column 을 한 번에 처리할 수 있다. 

```python
from sklearn.impute import SimpleImputer

# imputer 모델 선언 (결측치를 채우는 모델을 하나 가져와서 (mean으로 채우는))
imputer = SimpleImputer(strategy='mean')
```

```python
# fit
imputer.fit(train[['Age', 'Pclass']])
```

```python
# transform()
result = imputer.transform(train[['Age', 'Pclass']])
result
```

```python
train[['Age', 'Pclass']] = result
```

```python
train[['Age', 'Pclass']].isnull().sum()
```

```python
train[['Age', 'Pclass']].describe()
```

### 전처리 : Normalize (정규화) 

영화평점 리뷰  
네이버 평점 리뷰 (0점 ~ 10점) : .[2,4,6,8,10].  
넷플릭스 평점리뷰 (0-5점): .[1,2,3,4,5].  

두개의 데이터가 범위가 다르다 보니까 같은 형태의 데이터로 모아줌. 섞어 쓸 수 있게끔.   
10점 기준에 2,4,6~ 이나 5점 기준에 1,2,3~ 이나 같은것 이기 때문  

데이터간에 다른 min, max 값을 가지는 경우, 정규화를 통해 최소/최대치 값의 척도를 맞추어준다.  

```python
movie={'naver': [2,4,6,8,10],
        'netflix':[1,2,3,4,5]
       }
```

```python
import pandas as pd

movies = pd.DataFrame(data=movie)
movies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# 모델선언
# feature_range를 지정하면 min값과 max 값을 지정할 수 있다. 
min_max_scaler = MinMaxScaler(feature_range=(0,10))
```

```python
min_max_movie = min_max_scaler.fit_transform(movies)
```

```python
#feature_range가 default 일 경우 (지정하지 않을 경우)
# min=0, max=1. 정규분포가 0~1 사이로 바뀐것이다. 
pd.DataFrame(min_max_movie, columns = ['naver','netflix'])
```

현재 시간 값을 random state 지정해서 매순간 바뀜. 랜덤 스텟값을 내가 특정값을 지정. 그냥 씨드값이 다름. 결과가 다름. 씨드가 다르니까
